## 1.1 Data Description

This data set is giving information about some wide range of population on their occupation, educational qualification and their demographic details and how these conditions are affecting their income.

## 1.2 Objective

# 7. Feature Engineering 

## 7.1 Normalization for Numerical Features

best features:- marital-stautus, education-num, relationship, age

```python
df.head(20)
```

```python
df = df.drop(["workclass"], axis=1)
df = df.drop(["education"], axis=1)
df = df.drop(["marital-status"], axis=1)
df = df.drop(["occupation"], axis=1)
df = df.drop(["race"], axis=1)
df = df.drop(["gender"], axis=1)
df = df.drop(["capital-gain"], axis=1)
df = df.drop(["capital-loss"], axis=1)
df = df.drop(["native-country"], axis=1)

df.head()
```

```python
dummy = pd.get_dummies(df[['relationship']])
```

```python
print(dummy.head(20))
print(dummy.shape)
```

```python
df=df.merge(dummy, left_index=True, right_index=True)
df.head()
```

```python
df = df.drop(["relationship"], axis=1)

df.head()
```

```python
df['income'] = df['income'].map({'<=50K':0,'>50K':1})
df.head()
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, train_size=0.7, \
                                                    random_state=0)

print(df_train.shape)
print(df_test.shape)
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

rescale = StandardScaler()

num_features = ['age','educational-num','hours-per-week']

norm_train_data = rescale.fit_transform(df_train.loc[:,num_features])

print(norm_train_data.shape, type(norm_train_data))
```

```python
df_train_norm = pd.DataFrame(norm_train_data, columns = num_features)
df_train=df_train.merge(df_train_norm, left_index=True, right_index=True)
df_train.head()
```

```python
y_train = df_train.pop('income')
x_train = df_train
```

```python
print(x_train)
```

```python
print(y_train)
```

### Training

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

### Making the Prediction

```python
num_features = ['age','educational-num','hours-per-week']
norm_test_data = rescale.transform(df_test.loc[:,num_features])

print(norm_test_data.shape, type(norm_test_data))
```

```python
df_test_norm = pd.DataFrame(norm_test_data, columns = num_features)
df_test=df_test.merge(df_train_norm, left_index=True, right_index=True)
df_test.head()
```

```python
y_test = df_test.pop('income')
x_test = df_test
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
y_pred = classifier.predict(x_test)
```

## Measuring Performance
### Accuracy

```python
# calculate accuracy of class predictions
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

### Confusion Matrix

```python
# print the confusion matrix
con_metrics = metrics.confusion_matrix(y_test, y_pred)
con_metrics
```

```python
#ploting heatmap for confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(con_metrics, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

### Classification Report

```python
#Checking Precision, Recall and F1 Score
print(metrics.classification_report(y_test, y_pred))
# Number of features= 4{age, educational-num, hours-per-week, relationship}
```
